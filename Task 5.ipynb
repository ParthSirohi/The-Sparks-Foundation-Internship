# Explore Business Analytics 

Performing ‘Exploratory Data Analysis’ on the provided dataset
‘SampleSuperstore’

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('Datasets/SampleSuperstore.csv')
data.head()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Category',data=data, palette='rainbow')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Sub-Category',data=data, palette='rainbow')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Region',data=data, palette='rainbow')
```

```python
# Let's try to visualize the 'Region' column from the Shipmode column dataset standpoints.

region_hist = sns.FacetGrid(data, col='Ship Mode', palette='rainbow')
region_hist.map(plt.hist, 'Region')
region_hist.set_ylabels('Number')
```

```python
# Let's try to visualize the 'Category' column from the Shipmode column dataset standpoints.

category_hist = sns.FacetGrid(data, col='Ship Mode', palette='rainbow')
category_hist.map(plt.hist, 'Category')
category_hist.set_ylabels('Number')
```

```python
# Let's try to visualize the 'Category' column from the Shipmode column dataset standpoints.

pclass_age_grid = sns.FacetGrid(data, col='Category', row='Sub-Category', height=2.2, aspect=1.6)
pclass_age_grid.map(plt.hist, 'Segment')
pclass_age_grid.set_ylabels('Number')
pclass_age_grid.add_legend()
```
